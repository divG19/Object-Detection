```python
"""
Splitting image channels
"""

r, g, b = cv2.split(new_image)
print('r', r.shape)
print('g', g.shape)
print('b', b.shape)
```

```python
new_image = cv2.merge((r,g,b))
```

```python
"""
Resize the image
"""

s = 10
w = int(new_image.shape[1]*s/100)
h = int(new_image.shape[0]*s/100)
dim = (w,h)
re_size = cv2.resize(new_image, dim, interpolation = cv2.INTER_AREA)
re_size.shape
```

```python
plt.imshow(re_size)
```

```python
"""
Rotate
"""

(h, w) = new_image.shape[:2]

c = (w/2, h/2)

angle = 90

m = cv2.getRotationMatrix2D(c, angle, 1.0)
rotate_90 = cv2.warpAffine(new_image, m, (h,w))
```

```python
plt.imshow(rotate_90)
```

```python
img = cv2.imread("traffic_jam_1050x700.webp")

blob = cv2.dnn.blobFromImage(img, 1/255, (320,320), (0,0,0), swapRB=True, crop = False)
```

```python
blob.shape
```

```python
"""
To print image
"""

i = blob[0].reshape(320,320,3)
plt.imshow(i)
```

```python
yolo.setInput(blob)
```

```python
output_layer_name = yolo.getUnconnectedOutLayersNames()
layeroutput = yolo.forward(output_layer_name)
```

```python
boxes = []
confidence = []
class_ids = []
width = 320
height = 320

for output in layeroutput:
    for detection in output:
        score = detection[:5]
        class_id = np.argmax(score)
        confidence = score[class_id]
        if confidence > 0.7:
            center_x = int(detection[0]*width)
            center_y = int(detection[0]*height)
            w = int(detection[0]*width)
            h = int(detection[0]*height)
            
            x = int(center_x - w/2)
            y = int(center_y - h/2)
            
            boxes.append([x,y,w,h])
            confidence.append(confidence)
            class_ids.append(class_id)
```
